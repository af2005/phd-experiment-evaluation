```python
from gascooling.accommodation.evaluation.experiment import influx

data_tm = influx.InfluxDataContainer(start="2023-04-16T08:00:00Z", stop="2023-04-18T21:00:00Z")
data_pressure = influx.InfluxDataContainer(
    start="2023-04-16T08:00:00Z",
    stop="2023-04-18T21:00:00Z",
    field="Pressure_Gas_handling_(mbar)",
)
ax2 = (data_pressure.data.set_index("_time")["Pressure_Gas_handling_(mbar)"] * 10000).plot(
    ylabel="Pressure Gas handling (1e-4 mbar)",
    secondary_y=True,
)
ax1 = data_tm.data.set_index("_time")["Temperature_Testmass_(K)"].plot(
    secondary_y=False,
)

ax1.set_ylabel("Temperature Testmass (K)")
ax1.set_zorder(100)
ax1.set_facecolor("none")
```
