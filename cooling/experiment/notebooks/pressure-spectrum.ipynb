```python

from gascooling.accommodation.evaluation import experiment
import matplotlib.pyplot as plt
from scipy import signal
```

```python
data = experiment.influx.InfluxDataContainer(
    field="Pressure_Gas_handling_(mbar)",
    start="2023-04-19T18:00:00Z",
    stop="2023-04-23T12:00:00Z",
)
pressure = data.data["Pressure_Gas_handling_(mbar)"]
```

```python
display(pressure)
```

```python
freqs, psd = signal.welch(pressure)

plt.figure(figsize=(5, 4))
plt.semilogx(freqs, psd)
plt.title("PSD: power spectral density")
plt.xlabel("Frequency")
plt.ylabel("Power")
plt.tight_layout()
```

```python
f, Pxx_den = signal.periodogram(pressure, 1 / 30)

plt.semilogx(f, Pxx_den)


plt.xlabel("frequency [Hz]")

plt.ylabel("PSD [V**2/Hz]")

plt.show()
```
