```python
import matplotlib.pyplot as plt
import pandas as pd

import helpers

helpers.figures.initialize()

df = pd.read_csv(
    "reflectivity.csv",
    names=["Wavelength (nm)", "Reflectivity"],
    index_col="Wavelength (nm)",
)
df2 = pd.read_csv(
    "absorption.csv",
    names=["Wavelength (nm)", "Absorption"],
    index_col="Wavelength (nm)",
)


ax = df.plot.line(figsize=helpers.figures.set_size())
df2.plot.line(ax=ax)
ax.set_xlim([900, 1200])
plt.axvline(x=1064, color="gray", ls="--", label="1064")
ax.set_xticks([900, 950, 1000, 1064, 1100, 1150, 1200])
xlabel = ax.set_xlabel("Wavelength (nm)")
ylabel = ax.set_ylabel("Reflectivity, Absorption (%)")

plt.tight_layout()
ax.figure.savefig(
    "reflectivity.pgf",
    backend="pgf",
    bbox_extra_artists=[xlabel, ylabel],
    bbox_inches="tight",
)
```
