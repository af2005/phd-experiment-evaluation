```python
import matplotlib.pyplot as plt
import numpy as np

from cooling.theory.data import materials

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
temps = np.linspace(4, 300)
heats = materials.teflon.heat_conductivity(temps)
ax1.semilogx(temps, heats, label="Thermal conductivity")
ax1.set_xlabel("Temperature (K)")
ax1.set_ylabel("Heat conductivity (W/m-K)")

specific_heat = materials.teflon.specific_heat(temps)
ax2.semilogx(temps, specific_heat, label="Specific heat", color="C2")
ax2.set_ylabel("Specific heat (J/kg-K)")
fig.legend()
plt.tight_layout()
fig.savefig("thermal-properties.pdf")
```

```python
test_mass_temperature = 50
holder_temperature = 10
test_mass_holder_distance = 2e-3

screw_diameter = 1e-3
conductive_cooling = (
    (screw_diameter / 2) ** 2
    * np.pi
    * materials.teflon.heat_conductivity(holder_temperature)
    * (test_mass_temperature - holder_temperature)
    / test_mass_holder_distance
)
print(conductive_cooling)
```
